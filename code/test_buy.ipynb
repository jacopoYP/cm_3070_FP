```python
import numpy as np

X = np.load("data/features.npy")
p = np.load("data/prices.npy")

print(X.shape, p.shape)
print("features finite:", np.isfinite(X).all())
print("prices finite:", np.isfinite(p).all())

# check that z-scored features are roughly centered
print("feature means (approx):", X.mean(axis=0))
print("feature stds (approx):", X.std(axis=0))

# prices should be positive
print("min price:", p.min(), "max price:", p.max())
```
