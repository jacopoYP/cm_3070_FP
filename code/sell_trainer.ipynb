```python
import matplotlib.pyplot as plt

def plot_loss(agent):
    losses = agent.loss_history
    plt.figure(figsize=(10,5))
    plt.plot(losses, alpha=0.7)
    plt.title("DDQN Loss History (Buy Agent - Multiprocess Training)")
    plt.xlabel("Training Steps")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.show()
```
